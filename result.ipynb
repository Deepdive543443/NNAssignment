```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from utils import *

```

Load and plot the result saved from training.ipynb

```python
convs_curve_train, convs_curve_test, convs_curve_train_acc, convs_curve_test_acc = torch.load("result/Base_line_model.tar")
print(len(convs_curve_train))
```

```python
length = 100

print(f"Max accuracy:{max(convs_curve_test_acc)}   mininum loss:{min(convs_curve_test)}")
print(f"Max epoch:{np.argmax(convs_curve_test_acc)}")

plt.plot(convs_curve_train[:length], label="train_loss")
plt.plot(convs_curve_test[:length], label="test_loss")
draw_early_stop(convs_curve_test, "min loss")
plt.xlabel("Epochs")
plt.ylabel("Loss per item")
plt.title("Loss curve of proto model")
plt.legend()
plt.show()


plt.plot(convs_curve_train_acc[:length], label="train_acc")
plt.plot(convs_curve_test_acc[:length], label="test_acc")
draw_max_acc(convs_curve_test_acc,"Max acc")
plt.xlabel("Epochs")
plt.ylabel("Acc per item")
plt.title("Acc curve of proto model")
plt.legend()
plt.show()
```
